```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "student": ["Ali", "Sara", "John", "Amina", "Zara", None],
    "age": [20, 21, np.nan, 22, 21, 20],
    "attendance": [85, 90, 60, None, 95, 70],
    "marks": [88, 92, 65, 80, None, 72]
}

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["attendance"] = df["attendance"].fillna(df["attendance"].mean())
df["age"] = df["age"].fillna(df["age"].mean())
df["marks"] = df["marks"].fillna(df["marks"].mean())
df
```

```python

df.describe()
```

I have fill the unknown age attendace and marks by computing the average vallue of each column and adding it to the unknown values

```python
plt.scatter(df["attendance"], df["marks"])
plt.xlabel("Attendance")
plt.ylabel("Marks")
plt.title("Attendace vs Marks")
plt.show()
```

attendance shows a linear relation with marks.

age does not have a relation with marks..

```python
plt.hist(df["marks"], bins=10)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Distribution of Marks")
plt.show()
```

```python
plt.boxplot(df["marks"])
plt.title(" marks Boxplot ")
plt.show()
```
